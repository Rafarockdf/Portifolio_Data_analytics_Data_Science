```python
# Função de limpeza de valores nulos
def limpeza(df, estrategia='remover', valor_substituto=None):
  if estrategia == 'remover':
        df_sem_nulos = df.dropna()
        df_sem_nulos = df_sem_nulos.replace('"', ' ')
        return df_sem_nulos
  elif estrategia =='substituir':
        if valor_substituto is None:
            raise ValueError("Você precisa fornecer um valor de substituição.")
        df_substituido = df.fillna(valor_substituto)
        df_substituido = df_substituido.replace('"', ' ')
        return df_substituido
```
